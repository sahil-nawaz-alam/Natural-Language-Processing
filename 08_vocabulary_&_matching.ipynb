```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
from spacy.matcher import Matcher
matcher =Matcher(nlp.vocab)
```

```python
pattern1 = [{'LOWER':'solarpower'}]
pattern2 = [{'LOWER':'Solar'},{'LOWER':'Power'}]
pattern3 = [{'LOWER':'Solar'},{'IS_PUNCT':True},{'LOWER':'Power'}]

matcher.add('SolorPower' , [pattern1 , pattern2, pattern3])
```

```python
doc = nlp(u'The Solar Power industry continues to grow as demand \ for solarpower increases. Solar-Power cars are gaining popularity')
```

```python
found_matches = matcher(doc)
```

```python
found_matches
```

```python
for match_id , start , end in found_matches:
  string_id = nlp.vocab.strings[match_id]
  span = doc [start:end]
  print(match_id , string_id , start , end , span.text)
```

```python
pattern1 = [{'LOWER':'solarpower'}]
pattern2 = [{'LOWER':'Solar'},{'IS_PUNCT':True , 'OP':'*'},{'LOWER':'Power'}]

matcher.add('solorPower' , [pattern1 , pattern2])
```

```python
found_matche = matcher(doc)
found_matche
```

```python
from spacy.matcher import PhraseMatcher
matcher = PhraseMatcher(nlp.vocab)
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/NLP/reaganomics.txt' , 'r' , encoding='latin-1') as f:
  doc3 = nlp(f.read())
```

```python
pharse_list = ['voodoo economics' ,  'supply-side economics' , 'trickle-down econimics', 'free-market economics']

pharse_patterns = [nlp(text) for text in pharse_list]

matcher.add('voodooEconomics' , None , *pharse_patterns)

matches = matcher(doc3)
```
